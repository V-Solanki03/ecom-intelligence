```python
import pandas as pd

df = pd.read_csv("../data/fakestore_raw.csv")
df.head()
```

```python
df.info()
```

```python
df["rating"] = df["rating"].astype(str)

df["rating_rate"] = df["rating"].str.extract(r"'rate': ([0-9.]+)").astype(float)
df["rating_count"] = df["rating"].str.extract(r"'count': (\d+)").astype(int)

df.drop(columns=["rating"], inplace=True)
```

```python
df.isnull().sum()
```

```python
df["rating_rate"].fillna(df["rating_rate"].mean(), inplace=True)
df["rating_count"].fillna(0, inplace=True)
```

```python
df["category"] = df["category"].str.lower().str.strip()
```

```python
df["category"] = df["category"].replace({
    "men's clothing": "mens clothing",
    "women's clothing": "womens clothing"
})
```

```python
df.drop_duplicates(subset="id", inplace=True)
```

```python
def price_band(price):
    if price < 25:
        return "Budget"
    elif price < 75:
        return "Mid"
    else:
        return "Premium"

df["price_band"] = df["price"].apply(price_band)
```

```python
df.info()
df.head()
```

```python
df.to_csv("../data/fakestore_clean.csv", index=False)
print("✅ clean dataset saved")
```
